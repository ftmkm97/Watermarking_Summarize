```python
import pandas as pd
import numpy as np

df= pd.read_csv('Xsum_preprocessing.csv')
df
```

# Removing Punctuations

```python
import re

df["document_preproced"] = df["document"].apply(lambda s: ' '.join(re.sub("[.,!?:;-='...\"@#_]", " ", s).split()))
df
```

# Removing Numbers

```python
df["document_preproced"] = df["document_preproced"].replace('\d+', '', regex=True)
df
```

# Removing Emojies

```python
def deEmojify(inputString):
    return inputString.encode('ascii','ignore').decode('ascii')

df['document_preproced']= df['document_preproced'].apply(lambda s: deEmojify(s))
df
```

# Removing Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = stopwords.words('english')
```

```python
def rem_en(input_txt):
    words= input_txt.lower().split()
    noise_free = [word for word in words if word not in stop]
    noise_free = " ".join(noise_free)
    return(noise_free)


df['document_preproced'] = df['document_preproced'].apply(lambda s: rem_en(s))
df
```

# Stemming

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")


def stim_text(text):
  noise_free = [stemmer.stem(w) for w in text.split()]
  noise_free = " ".join(noise_free)
  return(noise_free)


df['document_preproced'] = df['document_preproced'].apply(lambda s: stim_text(s))

df
```

# Lemmating

```python
import nltk
nltk.download('wordnet')


w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    noise_free = [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
    noise_free = " ".join(noise_free)
    return(noise_free)


df['document_preproced'] = df['document_preproced'].apply(lambda s: lemmatize_text(s))

df
```

```python
df.to_csv('Xsum_preprocessing.csv',index=False)
```
